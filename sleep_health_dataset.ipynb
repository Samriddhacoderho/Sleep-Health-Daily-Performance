```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/content/sleep-health-and-daily-performance-dataset/sleep_health_dataset.csv')
```

```python
X=df.drop(['sleep_disorder_risk','person_id'],axis=1)
y=df['sleep_disorder_risk']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer,make_column_selector
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline

ordinal_pipeline = Pipeline([
    ('encode', OrdinalEncoder(categories=[['Depression','Both','Anxiety','Healthy']])),
    ('scale', StandardScaler())
])

transformer = ColumnTransformer(transformers=[
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first'), [1,2,4,19,26,27]),
    ('ordinal', ordinal_pipeline,[20]),
    ('num', StandardScaler(), make_column_selector(dtype_include=['int64', 'float64']))
])
```

```python
X_train=transformer.fit_transform(X_train)
X_test=transformer.transform(X_test)
```

```python
X_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
Y_train=encoder.fit_transform(Y_train)
Y_test=encoder.transform(Y_test)
```

```python
import torch
from torch.utils.data import Dataset,DataLoader


class CustomDataSet(Dataset):
  def __init__(self,features,labels):
    self.features=torch.tensor(features,dtype=torch.float32)
    self.labels=torch.tensor(labels,dtype=torch.long)

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,idx):
    return self.features[idx],self.labels[idx]


train_set=CustomDataSet(X_train,Y_train)
test_set=CustomDataSet(X_test,Y_test)

train_loader=DataLoader(train_set,batch_size=64,shuffle=True)
test_loader=DataLoader(test_set,batch_size=64,shuffle=False)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
#Model Creation

class ANNModel(nn.Module):
  def __init__(self,input_features,output_features):
    super().__init__()
    self.network=nn.Sequential(
        nn.Linear(input_features,128),
        nn.ReLU(),
        nn.Linear(128,64),
        nn.ReLU(),
        nn.Linear(64,32),
        nn.ReLU(),
        nn.Linear(32,output_features)
    )

  def forward(self,x):
    return self.network(x)
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
#Train Loop

model=ANNModel(X_train.shape[1],df['sleep_disorder_risk'].value_counts().shape[0])
model.to(device)
criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)


for i in range(50):
  tot_loss=0
  for x,y in train_loader:
    x=x.to(device)
    y=y.to(device)
    y_pred=model(x)
    loss=criterion(y_pred,y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    tot_loss+=loss.item()
  print(f'Epoch {i+1}, Avg Loss: {tot_loss/len(train_loader)}')
```

```python
model.eval()
```

```python
total=0
correct=0
with torch.no_grad():
  for x,y in test_loader:
    x=x.to(device)
    y=y.to(device)
    y_pred=model(x)
    _,predicted=torch.max(y_pred,1)
    total=total+y.shape[0]
    correct=correct+(predicted==y).sum().item()
print(f'Accuracy: {correct/total}')
```
