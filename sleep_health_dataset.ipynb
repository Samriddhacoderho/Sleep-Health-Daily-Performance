```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer,make_column_selector
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline

ordinal_pipeline = Pipeline([
    ('encode', OrdinalEncoder(categories=[['Depression','Both','Anxiety','Healthy']])),
    ('scale', StandardScaler())
])

transformer = ColumnTransformer(transformers=[
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first'), [1,2,4,19,26,27]),
    ('ordinal', ordinal_pipeline,[20]),
    ('num', StandardScaler(), make_column_selector(dtype_include=['int64', 'float64']))
])
```

```python
X_train=transformer.fit_transform(X_train)
X_test=transformer.transform(X_test)
```

```python
X_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
Y_train=encoder.fit_transform(Y_train)
Y_test=encoder.transform(Y_test)
```

```python
import torch
from torch.utils.data import Dataset,DataLoader


class CustomDataSet(Dataset):
  def __init__(self,features,labels):
    self.features=torch.tensor(features,dtype=torch.float32)
    self.labels=torch.tensor(labels,dtype=torch.long)

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,idx):
    return self.features[idx],self.labels[idx]


train_set=CustomDataSet(X_train,Y_train)
test_set=CustomDataSet(X_test,Y_test)

```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
#Model Creation

class ANNModel(nn.Module):
  def __init__(self,input_dim,output_dim,num_of_layers,num_of_nodes_per_layer,dropout_size):
    super().__init__()
    layers=[]
    for i in range (num_of_layers):
      layers.append(nn.Linear(input_dim,num_of_nodes_per_layer))
      layers.append(nn.BatchNorm1d(num_of_nodes_per_layer))
      layers.append(nn.ReLU())
      layers.append(nn.Dropout(dropout_size))
      input_dim=num_of_nodes_per_layer

    layers.append(nn.Linear(num_of_nodes_per_layer,output_dim))
    self.network=nn.Sequential(*layers)


  def forward(self,x):
    return self.network(x)
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
!pip install optuna
import optuna
```

```python
import optuna.logging

optuna.logging.set_verbosity(optuna.logging.INFO)
```

```python
def objective(trial):
  #Hyper parameter define
  print(f"\n Starting Trial {trial.number}")
  num_of_layers=trial.suggest_int('num_of_layers',2,20,step=2)
  num_of_nodes_per_layer=trial.suggest_int('num_of_nodes_per_layer',4,128,step=8)
  dropout_size=trial.suggest_float('dropout_size',0.1,0.9,step=0.1)
  num_of_epochs=trial.suggest_int('num_of_epochs',20,200,step=20)
  learning_rate=trial.suggest_float('learning_rate',1e-5,1e-1,log=True)
  optimizer_name=trial.suggest_categorical('optimizer_name',['Adam','SGD','RMSprop'])
  batch_size=trial.suggest_categorical('batch_size',[8,16,32,64,128])
  weight_decay=trial.suggest_float('weight_decay',1e-5,1e-1,log=True)

  #DataLoader Creation

  train_loader=DataLoader(train_set,batch_size=batch_size,shuffle=True,pin_memory=True)
  test_loader=DataLoader(test_set,batch_size=batch_size,shuffle=False,pin_memory=False)

  #Model Object Creation
  model=ANNModel(X_train.shape[1],df['sleep_disorder_risk'].value_counts().shape[0],num_of_layers,num_of_nodes_per_layer,dropout_size)
  model.to(device)


  #Optimizer and Loss Function Selection
  criterion=nn.CrossEntropyLoss()
  if(optimizer_name=="Adam"):
    optimizer=optim.Adam(model.parameters(),lr=learning_rate,weight_decay=weight_decay)
  elif(optimizer_name=="SGD"):
    optimizer=optim.SGD(model.parameters(),lr=learning_rate,weight_decay=weight_decay)
  else:
    optimizer=optim.RMSprop(model.parameters(),lr=learning_rate,weight_decay=weight_decay)


  #Training Loop
  for i in range(num_of_epochs):
    for x,y in train_loader:
      x=x.to(device)
      y=y.to(device)
      y_pred=model(x)
      loss=criterion(y_pred,y)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()


  #Evaluation Mode
  model.eval()
  total=0
  correct=0

  #Test Loop
  with torch.no_grad():
    for x,y in test_loader:
      x=x.to(device)
      y=y.to(device)
      y_pred=model(x)
      _,predicted=torch.max(y_pred,1)
      total=total+y.shape[0]
      correct=correct+(predicted==y).sum().item()

  print(f"Trial {trial.number} Accuracy: {correct/total}")
  return (correct/total)

```

```python
study=optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=100)
```
